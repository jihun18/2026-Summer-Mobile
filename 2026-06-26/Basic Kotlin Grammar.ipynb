## 변수 선언 학습
- var: 변수, val: 상수 선언
- 엄격한 데이터 타입 지정 필요: Double, Float에 Int 객체 지정 불가

```kotlin
var data1 : Double = 20.0
// var data1:Double = 20 오류 // Int를 Double로 변환하지 못해서 오류
// var data1:Double = 20. 오류 // 숫자 다음에 .만 있어서 오류
var data2 = 20.3
var data3 = '가'

println("$data1, $data2, $data3")
// data1 = 30 오류 // 정수 객체를 Double 객체로 변환해서 오류
data1 += 30
data1 = data1 + 20
data1 += 15.0
println("$data1, $data2, $data3")

var data4:Float = 25.0F // xxxF 숫자 형식 필요
println("Float: $data4")

var data5:Long = 13L // L은 선택
println("Long: $data5")

var data6:Short = 777 // 숫자의 크기(범위) 확인 필요
```

## 함수와 데이터 타입
- 반환문이 없는 경우: Unit 사용
- Unit 반환문 생략 가능

```kotlin
fun add(a: Int, b: Int): Int{
    return a+b
}

println("2 + 3 = ${add(2,3)}")

fun showMessage(mesg: String){
    println(mesg)
}

showMessage("안녕하세요")
```

### 함수의 기본 값 정의
- 함수 인수에 값을 지정하면, 가변 인수를 전달할 수 있음

```kotlin
import kotlin.math.pow
fun power(base:Double, exp:Int = 2): Double {
    return base.pow(exp)
}

println("2^3 = ${power(2.0, 3)}")
println("3^2 = ${power(3.0)}")
```

## if 제어문
- 일반 프로그래밍 언어와 동일
- 삼항연산자처럼 if 조건식에 따라 값을 지정 가능

```kotlin
import kotlin.random.Random
val score = Random.nextInt(0, 101) // 0~100까지
val cut_off = if(score >= 60) {
    println("점수를 비교해서, String 합격 여부를 반환.")
    "Dummy"
    "Pass" }
     else "Fail"
println("너의 점수는 ${score}점이라서 ${cut_off}야.")
```

## when 제어문 사용
- when 제어문은 기존의 switch문보다 강력함
- 일반적인 switch문은 if 동등절에 해당하나, when 제어문은 범위 및 다양한 조건을 포함할 수 있음

```kotlin
//println("입력 스트링:")
//val input: String = readln()
val data: Any = "Test"
when(data) {
    is String -> println("입력이 String임.")
    20, 30 -> println("입력 값이 20 또는 30임")
    in 1..10 -> println("입력 값이 1 이상 10 이하") 
    else -> println("입력 값이 예상 범위를 벗어남")
}
```

## 반복문을 통해서 원주율 계산하기
- pi/4 = 1 - 1/3 + 1/5 - 1/7 + 1/9 - 1/11 + 1/13 ...

```kotlin
fun series(n: Int): Double {
    var x: Double = 0.0
    
    for (i in 1..n) {
        val denom = 2 * i - 1
        val sign = if (i % 2 == 1) 1 else -1
        x += sign / denom.toDouble()
    }

    return x * 4
}

println("Series(100) = ${series(100)}")
println("Series(1000) = ${series(1000)}")
println("Series(50000) = ${series(50000)}")
```
